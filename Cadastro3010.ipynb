```python
import sqlite3

conn = sqlite3.connect('empresa.db')
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS funcionarios (
    id INTEGER NOT NULL PRIMARY KEY AUTOINCREMENT,
    nome TEXT NOT NULL,
    cargo TEXT NOT NULL,
    salario REAL NOT NULL
);
""")
conn.commit()

def adicionar_funcionario(nome, cargo, salario):
  """Adiciona um novo funcionário à tabela funcionarios."""
  cursor.execute("""
  INSERT INTO funcionarios (nome, cargo, salario)
  VALUES (?, ?, ?)
  """, (nome, cargo, salario))
  conn.commit()
  print(f"Funcionário {nome} adicionado com sucesso!")

def listar_funcionarios():
  """Lista todos os funcionários presentes na tabela."""
  cursor.execute("SELECT id, nome, cargo, salario FROM funcionarios")
  funcionarios = cursor.fetchall()

  if not funcionarios:
    print("Não há funcionários na tabela.")
  else:
    print("\n--- Lista de Funcionários ---")
    for funcionario in funcionarios:
      print(f"ID: {funcionario[0]}, Nome: {funcionario[1]}, Cargo: {funcionario[2]}, Salário: {funcionario[3]}")
    print("-----------------------------\n")


def buscar_funcionario_por_nome(nome):
  """Busca funcionários na tabela pelo nome."""
  cursor.execute("SELECT id, nome, cargo, salario FROM funcionarios WHERE nome = ?", (nome,))
  funcionarios_encontrados = cursor.fetchall()

  if not funcionarios_encontrados:
    print(f"Nenhum funcionário encontrado com o nome: {nome}")
  else:
    print(f"\n--- Funcionários encontrados com o nome: {nome} ---")
    for funcionario in funcionarios_encontrados:
      print(f"ID: {funcionario[0]}, Nome: {funcionario[1]}, Cargo: {funcionario[2]}, Salário: {funcionario[3]}")
    print("-----------------------------------------------------\n")


def calcular_media_salarial():
  """Calcula a média salarial de todos os funcionários na tabela."""
  cursor.execute("SELECT AVG(salario) FROM funcionarios")
  average_salary = cursor.fetchone()[0]

  if average_salary is not None:
      print(f"A média salarial dos funcionários é: {average_salary:.2f}\n")
  else:
      print("Não foi possível calcular a média salarial.")

while True:
  print("\nMenu:")
  print("1. Adicionar funcionário")
  print("2. Listar todos os funcionários")
  print("3. Buscar funcionário por nome")
  print("4. Calcular média salarial")
  print("5. Sair")

  choice = input("Escolha uma opção: ")

  try:
    choice = int(choice)

    if choice == 1:
      nome = input("Digite o nome do funcionário: ")
      cargo = input("Digite o cargo do funcionário: ")
      while True:
          salario_str = input("Digite o salário do funcionário: ")
          try:
              salario = float(salario_str)
              break
          except ValueError:
              print("Entrada inválida para salário. Por favor, digite um número.")
      adicionar_funcionario(nome, cargo, salario)
    elif choice == 2:
      listar_funcionarios()
    elif choice == 3:
      nome = input("Digite o nome do funcionário para buscar: ")
      buscar_funcionario_por_nome(nome)
    elif choice == 4:
      calcular_media_salarial()
    elif choice == 5:
      print("Saindo do programa.")
      break
    else:
      print("Opção inválida. Por favor, escolha um número entre 1 e 5.")
  except ValueError:
    print("Entrada inválida. Por favor, insira um número.")

conn.close()
print("Conexão com o banco de dados fechada.")
```
